```python

features = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies', 'PaperlessBilling'
]
feature_counts = {}

for feature in features:
  if feature == 'InternetService':
      feature_counts[feature] = (df[feature] != 'No').sum()
  else:
      feature_counts[feature] = (df[feature] == 'Yes').sum()


plt.bar(feature_counts.keys(), feature_counts.values())

plt.xlabel('Features')
plt.ylabel('Count')
plt.title('Total count of Features Being used by customers')


plt.xticks(rotation=45)

plt.show()

```

```python
import numpy as np


gender_labels = ['Male', 'Female']
churn_labels = churn_gender.index.levels[0]
x = np.arange(len(gender_labels))
width = 0.35

fig, ax = plt.subplots()
for i, churn_label in enumerate(churn_labels):
    counts = churn_gender[churn_label]
    ax.bar(x + i * width, counts, width, label=churn_label)

ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Distribution of Gender Based on Churn Status')
ax.set_xticks(x + width * len(churn_labels) / 2)
ax.set_xticklabels(gender_labels)
ax.legend(title='Churn')

plt.show()






```

```python
yeschurn_gender_seniorcitizen = df[df['Churn']=='Yes'].groupby('gender')['SeniorCitizen'].value_counts()
print(yeschurn_gender_seniorcitizen)
```

```python
import matplotlib.pyplot as plt

gender_labels = yeschurn_gender_seniorcitizen.index.get_level_values('gender')
senior_citizen_values = yeschurn_gender_seniorcitizen.index.get_level_values('SeniorCitizen')

labels = []
for gender, senior in zip(gender_labels, senior_citizen_values):
    senior_label = 'Non senior citizen' if senior == 0 else 'Senior citizen'
    labels.append(f'{gender} - {senior_label}')

sizes = yeschurn_gender_seniorcitizen.values


plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])


plt.axis('equal')


plt.title('Proportion of Senior Citizens Churned (Churn = Yes) Based on Gender')

plt.show()





```

```python
internet = df[df['InternetService'] != 'No'].groupby('InternetService')['OnlineSecurity'].apply(lambda x: (x == 'Yes').sum())
print(internet)
```

```python
labels = internet.index
sizes = internet.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen'])

plt.axis('equal')
plt.title('Proportion of Customers with Online Security Based on Internet Service')

plt.show()
```

```python
columns = ['OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

plots = []


fig, axes = plt.subplots(1, len(columns), figsize=(15, 5))

for ax, column in zip(axes, columns):
    counts = df[df['InternetService'] != 'No'].groupby('InternetService')[column].apply(lambda x: (x == 'Yes').sum())
    labels = counts.index
    sizes = counts.values

    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
    ax.axis('equal')
    ax.set_title(f'Distribution of {column}')

    plots.append(ax)


plt.tight_layout()
plt.show()
```
